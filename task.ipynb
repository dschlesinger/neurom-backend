```python
import pandas as pd, matplotlib.pyplot as plt, random, \
    json
```

```python
df = pd.read_csv('EEG_recording_2025-10-27-23.02.20.csv')
```

```python
sensors = ['TP9', 'AF7', 'AF8', 'TP10', 'Right AUX']
```

```python
df[sensors].min()
```

```python
fig, axs = plt.subplots(5, 1)

axs = axs.flatten()

for s, ax in zip(sensors, axs):

    ax.plot(df[s].to_numpy())
```

```python
gathered = []
```

```python
width = 250

for i in range(0, df.__len__() -  width):

    section = df[sensors].to_numpy()[i : i + width, :]
    print(section.shape)

    if abs(section).max() < 50:

        gathered.append(section)
```

```python
len(gathered)
```

```python
examples = []

n: int = 50

for _ in range(n):

    examples.append(
        gathered.pop(
            random.randint(0, len(gathered) - 1)
        )
    )
```

```python
for e in examples:

    fig, axs = plt.subplots(5, 1)

    axs = axs.flatten()

    for i, ax in enumerate(axs):

        ax.plot(e[:, i])

plt.show()
```

```python
dps = [
    {
        'anom': {
            'start': 0,
            'end': 1,
            'data': d.tolist(),
            "sensors": [
                "TP9",
                "AF7",
                "AF8",
                "TP10",
                "Right AUX"
            ],
            "final": True
        },
        "classification": "Nothing"
    }
    for d in examples
]
```
